- We learned that the temperature at New Haven, Connecticut is currently 58 degree fahrenheit. Add this information into the DataFrame.

```python
#### Your code here
new_data = pd.DataFrame({'City': ["New Haven"], "State": ["Connecticut"], "Temperature": [58]})
temp = pd.concat([temp, new_data])

temp.reset_index()
```

- Replace the names of the states by their abbreviations. That is 
    - Connecticut --> CT
    - New Jersey  --> NJ
    - New York    --> NY

```python
#### Your code here
temp.replace({'Connecticut': 'CT', 'New Jersey': 'NJ', 'New York': 'NY'}, inplace=True)
temp
```

- Select the rows in New York.

```python
#### Your code here
ny_rows = temp[temp['State'] == 'NY']
```

- What is the average of the temperature of cities in NY?

```python
#### Your code here
ny_rows['Temperature'].mean()
```

```python
#### Your code here
results = temp.groupby('State')['Temperature'].agg(['mean', 'std'])
```
